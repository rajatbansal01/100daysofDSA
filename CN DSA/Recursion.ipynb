```python
# Sum of n natural numbers
def n_sum(n):
    if n == 0:
        return(0)
    else:
        return(n+n_sum(n-1))
```

```python
n_sum(0)
```

```python
# Power of a number

def pow(x,n):
    if n == 0:
        return(1)
    else:
        return(x*pow(x,n-1))
    
    
```

```python
pow(2,5)
```

```python
# fibonacci Series
# sum of previous two numbers
# 1,1,2,3,5,8,.....
```

```python
def fib(n):
    if n==1:
        return 1
    return fib(n-1) + fib(n-2)
```

```python
%%time
fib(5)
```

```python
## correct version:
def fib(n):
    if n==1 or n==2:
        return 1
    return fib(n-1) + fib(n-2)
```

```python
fib(5)
```

```python
# check if a list is sorted or not using recursion
```

```python
def check_list(a):
    l = len(a)
    if l==0 or l==1:
        return True
    if a[0]>a[1]:
        return False
#     smallerList = a[1:]
    return(check_list(a[1:]))
```

```python
check_list([1,2,3,4,5])
```

```python
# updated with list and start index
def isSorted(a, si):
    l = len(a)
    if  si == l-1 or si == l:
        return True
    if a[si]>a[si+1] :
        return False
    return isSorted(a,si+1)
```

```python
isSorted([1,2,30,4,5], 0)
```

### Assignment
Sum Of Array

Send Feedback

Given an array of length N, you need to find and return the sum of all elements of the array.

Do this recursively.

Input Format :

Line 1 : An Integer N i.e. size of array

Line 2 : N integers which are elements of the array, separated by spaces

Output Format :

Sum
Constraints :

1 <= N <= 10^3

Sample Input 1 :

3

9 8 9

Sample Output 1 :

26

Sample Input 2 :

3

4 2 1

Sample Output 2 :

7    


```python
def sum_arr(a):
    l = len(a)
    if l == 0:
        return 0
    return(a[0]+sum_arr(a[1:]))
```

```python
sum_arr([1,2,3,4,5, 6,7])
```

```python
# Better Approach ; no slicing
def sum_arr(a,si):
    l = len(a)
    if si == l-1:
        return a[si]
    return a[si] + sum_arr(a,si+1)
```

```python
sum_arr([1,2,3,4,5,6],0)
```
